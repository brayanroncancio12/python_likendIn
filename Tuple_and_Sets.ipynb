# Sets

```python
mySet = {"apple", "banana", "cherry"} # This is a set in Python, we define this 
                                      # using curly braces and separating the items with commas.
mySet
```

```python
mySet = set(("apple", "banana", "cherry")) # We can also create a set using the set() constructor.
mySet
```

### Properties of Sets
- Declared with curly brackets {}
- All elements are unique 
- The order doesn't matter 

```python
myList = ['a', 'b', 'b', 'c', 'c']
myList = set(myList) 
myList = list(set(myList))# We can convert a list to a set to remove duplicates, and then convert it back to a list.
myList
```

```python
myList[0] # This will give us an error because sets do not support indexing.
```

```python
number = 1 
1[0] # This will also give us an error because integers do not support indexing.
```

```python
mySet.add("orange") # We can add an item to a set using the add() method.
mySet
```

```python
'apple' in mySet # We can check if an item is in a set using the 'in' keyword.
```

```python
'z' in mySet # This will return False because 'z' is not in the set.
```

```python
len(mySet) # We can get the number of items in a set using the len() function.
```

```python
while len(mySet):
    print(mySet.pop()) # We can remove and return an arbitrary item from the set using the pop() method.
```

# Tuples

```python
myTuple = ('a', 'b', 'c') # This is a tuple in Python, 
                        #we define this using parentheses and separating the items with commas.
myTuple
```

```python
myTuple[0] # We can access items in a tuple using indexing, just like with lists. 
```

```python
myTuple[0] = 'x' # This will give us an error because tuples are immutable, 
                # meaning we cannot change their items after they have been created.
```

```python
def returnsMultipleValues():
    return 1, 2, 3 # A function can return multiple values as a tuple.

type(returnsMultipleValues()) # The return value of the function is a tuple.
```

```python
myTuple = (1, 2, 3) # We can also create a tuple using parentheses and separating the items with commas.
type(myTuple)
```

```python
a, b, c = myTuple # We can unpack a tuple into separate variables.
```

```python
print(a) # This will print 1
print(b) # This will print 2
print(c) # This will print 3
#
```

### Tuples 
- Declared with parentheses 
- Are ordered and "subscritable"
- Cannot be modified

### Why use Tuples 
- More efficient than list 
- They don't grow or change 
- Store compactly in memory
