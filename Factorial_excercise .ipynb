```python
def factorial(num):
    if type(num) == str:
        return None
    if num < 0 or num != int(num):
        return None
    if num == 0:
        return 1
    else:
        factorial = 1
        for i in range(1, num + 1):
            factorial *= i
        return factorial
```

```python
number = 5
result = print(factorial(number))
```
