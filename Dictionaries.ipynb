# Dictionaries


```python
from collections import defaultdict # import defaultdict to create a dictionary that provides a default value for non-existent keys
```

```python
animals = {
    'a' : 'aardvark',
    'b' : 'bear',
    'c' : 'cat',
}

animals
```

```python
animals['a']
```

```python
animals['d'] = 'dog' # add a new key-value pair to the dictionary
```

```python
animals
```

```python
animals['a'] = 'antilope' # change the value associated with the key 'a'
animals
```

```python
animals.keys() # get a list of all the keys in the dictionary
```

```python
animals.values() # get a list of all the values in the dictionary
```

```python
list(animals.keys()) # convert the keys to a list
```

```python
animals['e'] # this will raise a KeyError because 'e' is not a key in the dictionary
```

```python
animals.get('a') # this will return 'antilope' because 'a' is a key in the dictionary
```

```python
len(animals) # get the number of key-value pairs in the dictionary
```

```python
animals = {
    'a' : ['aardvark', 'anatelope'],
    'b' : ['bear'],
}
# This dictionary has lists as values, which allows us to store multiple animals for each letter.
```

```python
animals['b'].append('buffalo') # add 'buffalo' to the list of animals associated with the key 'b'
```

```python
animals
```

```python
animals['c'] = ['cat'] # add a new key 'c' with a list containing 'cat' as its value
```

```python
if 'c' not in animals:
    animals['c'] = [] # if 'c' is not a key in the dictionary, add it with an empty list as its value

animals['c'].append('cougar') # add 'cougar' to the list of animals associated with the key 'c'
```

```python
animals
```

## the default dict

```python
animals = defaultdict(list) # create a defaultdict with list as the default factory
```

```python
animals 
```

```python
animals['a'].append('aardvark') # add 'aardvark' to the list of animals associated with the key 'a'
animals
```

```python
animals['a'].append('elephant') # add 'aardvark' to the list of animals associated with the key 'a'
animals
```

```python
animals['f']  # access the value associated with the key 'f', which does not exist, so it will return an empty list as the default value
```
