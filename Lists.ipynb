# Lists

## List Slicing 

```python
myList = [1, 2, 3, 4, 5]
myList[3:] # this will return [4, 5] because it starts from index 3 and goes to the end of the list
```

```python
myList[0:6:2] # this will return [1, 3, 5] because it starts from index 0, goes up to index 6 (which is out of range,\
                #so it goes to the end of the list), and takes every 2nd element
```

```python
myList[0 :6: 3]
```

```python
myList[ : : 2]
```

```python
for i in range (100):
    print(i)
```

```python
myList = list(range(100))
```

```python
myList[ : :10]
```

```python
myList[ : :-20]
```

# Modifying Lists

```python
myList = [1, 2, 3, 4]
myList.append(5) # this will add 5 to the end of the list, so myList will become [1, 2, 3, 4, 5]
print(myList)
```

```python
myList.insert( 3, 'a new value')
print(myList) # this will insert 'a new value' at index 3, so myList will become [1, 2, 3, 'a new value', 4, 5]
```

```python
myList.remove('a new value') # this will remove 'a new value' from the list, so myList will become [1, 2, 3, 4, 5]
```

```python
myList
```

```python
myList.pop() # this will remove the last element from the list and return it, so it will return 5 and myList will become [1, 2, 3, 4]
```

```python
myList
```

```python
while len(myList):
    print(myList.pop())
```

```python
myList
```

```python
a = [1, 2, 3]
b = a
a.append(4)
print(b) # this will print [1, 2, 3, 4] because b is a reference to the same list as a, so when we modify a, we also modify b
```

```python
a = [1, 2, 3]
b = a.copy() # this will create a new list that is a copy of a, so b will be [1, 2, 3] but it will be a different list in memory
a.append(4)
print(a) # this will print [1, 2, 3, 4] because we modified a
print(b)
```
